```python
import random
from model.city import City
from controller.gaplot import genetic_algorithm_plot
import matplotlib.pyplot as plt
```

```python

def generate_city_list(num):
    return [City(x=random_position(), y=random_position()) for i in range(num)]

def random_position():
    return int(random.random() * 200)
  
CITY_NUMBERS = 10
cityList = generate_city_list(CITY_NUMBERS)
```

```python
tsp_data = {
    "population": cityList,
    "pop_size": 100,
    "elite_size": 20,
    "mutation_rate": 0.2,
    "generations": 500,
}
cities = genetic_algorithm_plot(tsp_data)
```

```python
from selection import rank_routes

cities

best_route_index = rank_routes(cities)[0][0]
best_route = cities[best_route_index]
```

```python

x_coords = [city.get_x for city in best_route]
y_coords = [city.get_y for city in best_route]

# Plot
plt.scatter(x_coords, y_coords, color="blue", label="Cities")
plt.plot(x_coords, y_coords, linestyle="--", color="gray", label="Path")

# Annotate each city
for i, city in enumerate(best_route):
    plt.text(city.get_x, city.get_y, f"City {i+1}")

# Labels and legend
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Cities Representation")
plt.legend()
plt.grid(True)
plt.show()
```

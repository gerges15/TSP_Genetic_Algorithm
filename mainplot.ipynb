```python
import random
from city import City
from gaplot import genetic_algorithm_plot

```

```python

def generate_city_list(num):
    return [City(x=random_position(), y=random_position()) for i in range(num)]

def random_position():
    return int(random.random() * 200)
  
CITY_NUMBERS = 20
cityList = generate_city_list(CITY_NUMBERS)
```

```python
tsp_data = {
    "population": cityList,
    "pop_size": 100,
    "elite_size": 20,
    "mutation_rate": 0.01,
    "generations": 500,
}
genetic_algorithm_plot(tsp_data)
```
